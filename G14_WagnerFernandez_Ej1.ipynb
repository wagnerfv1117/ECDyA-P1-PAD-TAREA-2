<p style="text-align: justify;">En el ámbito del desarrollo, los servicios REST y las APIs RESTful han transformado significativamente la manera en que las aplicaciones interactúan y comparten información, comprender estos conceptos es esencial para cualquier desarrollador o profesional dentro del contexto data, busque construir aplicaciones eficientes y escalables; este trabajo abordará de manera detallada las siguientes temáticas fundamentales, tenemos al REST Web Service, que es una arquitectura diseñada para la creación de servicios web, que facilita la comunicación entre sistemas mediante el protocolo HTTP. La adopción generalizada de las RESTful APIs se debe a su facilidad de uso y su capacidad para integrarse con diversas aplicaciones y servicios, proporcionando una forma clara y estructurada de interactuar con los datos y funcionalidades de una aplicación.</p>

<p style="text-align: justify;">Por ultimo se tiene el formato JSON (JavaScript Object Notation) es un formato de texto ligero que se utiliza para el intercambio de datos. #### 1. Extracción de la información

Desarrolle un código de python que le permita obtener los datos de las naves de SpaceX, desde la API spacexdata v4: https://api.spacexdata.com/v4/ships, tenga en cuenta que debe usar el paquete requests, funciones:

* request.get
* Response.json

   De ser necesario, puede revisar su documentación en:https://requests.readthedocs.io/en/latest/api/?requests.Request

<p style="text-align: justify;">Se procede con importar la librería requests, se utiliza para realizar peticiones HTTP de forma sencilla y eficiente, ya que permite interactuar con servidores web y APIs, conforme a lo consultado en el material de Idris (2014)</p>

```python
import requests as rq
```

```python
data = rq.request('GET', 'https://api.spacexdata.com/v4/ships')# obtener data con el metodo request.get
data
```

#### 2. Cargar o transformación de datos

Cargue el json de la respuesta del request anterior como un DataFrame de pandas (ayuda: función pd.DataFrame.from_dict).

Se procede con cargar los datos de la API y convertirlos a formato .JSON

```python
to_json = data.json()#conversión de datos a .json o diccionario
to_json
```

Luego, se procede con importar la librería pandas, que sirve para transformar el archivo anterior, que está en formato .JSON o diccionario a un DataFrame, utilizando el metodo _pd.DataFrame.from_dict_.

```python
import pandas as pd

df = pd.DataFrame.from_dict(to_json)# metodo para convertir en dataframe
```

Imprimir las ultimas 5 filas del DataFrame con el metodo .tail() de la librería Pandas

#### 4. Analisis de variables

Con la información anterior, en una celda markdown concluya cuales de las variables podrían ser utilizadas para un análisis y clasifíquelas entre variables cualitativas (categóricas) y cuantitativas (numéricas).

Para responder el interrogante, se utilizará un método con la librería pandas llamado dtypes(), determinando cuales son las variables numéricas y categóricas teniendo en cuenta el material del libro de Grus (2023):

```python
"""Metodo para dentificar variables numéricas"""
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns

"""Método para dentificar variables categóricas"""
columnas_categoricas = df.select_dtypes(include='object').columns

print("Las columnas numéricas del DataFrame son las siguientes:", columnas_numericas)
print("Las columnas categóricas del DataFrame son las siguientes:", columnas_categoricas)
```

Datos ausentes en el DataFrame

Se procede con determinar las columnas que tienen datos ausentes o nulos, utilizando el metodo .isnull() de la librería Pandas.

```python
values_null = df.isnull()#verificar valores nulos
values_null
```

Conteo de valores faltantes

Entre muchos metodos, segun lo consultado en el libro de cccc, se usa el metodo conocido como .isna().sum para responder el interrogante

```python
print("Conteo de valores ausentes:")

print('###################################################')


missing_data_per_column = df.isna().sum()#aquí se cuentan los valores ausentes en cada columna

print(missing_data_per_column)

print('###################################################')


total_datos_ausentes = df.isna().sum().sum()#aquí se cuentan valores ausentes en todo el DataFrame

print("Total de datos ausentes en el DataFrame es de:",total_datos_ausentes)
```

#### 5. Conteo de datos

Determine el total de naves y cuántas están activas.


* Se procede con hacer el conteo de naves, utilizando la librería Pandas, se crea una columna para clasificar los valores "VERDADERO" o "FALSO" con el metodo .replace()
* Se desarrolla una variable,donde cuente los datos que tengan las palabra "VERDADERO" y los datos que tengan la palabra "FALSO"

```python
df["conteo de naves"] = df["active"].replace({True: "VERDADERO", False: "FALSO"})#columna para clasificar como "VERDADERO" o "FALSO"

active_count = df["conteo de naves"].value_counts()#número de registros con "VERDADERO" y "FALSO"

print(active_count)
```

#### 6. Gráfico tipo histograma

Visualice el histograma de los pesos de las naves en kg.

Para visualizar el histograma, se procede con visualizar los datos de la columna con el método hist() de la librería Maplotlib, aplicandolo a la columna "mass_kg"

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))# tamaño de la visualización
plt.hist(df["mass_kg"], bins=20, edgecolor='black')
plt.xlabel("Pesos de las naves o barcos (kg)")
plt.ylabel("Frecuencia de datos")
plt.title("Grafico No. 1 - Histograma de los pesos de las naves - 'mass_kg'")
plt.grid(True)
plt.show()
```

#### 7. Adicional o _Bonus Track_ con el histograma 

<p style="text-align: justify;">Observando la grafica anterior, se analiza que en la columna "mass_kg" hay valores faltantes, suscitando un sesgo o asimetría, por lo cual como adicional, se procede con imputar la columna, utilizando la mediana de las variables consignada en la columna, desarrollando un procedimiento llamado ingeniería de caracterísrticas o "feature engineering" que implica la extracción y transformación de variables a partir de datos sin procesar. Zheng y Casari (2018)</p>

```python
df_imputed = df['mass_kg'].median()#calcular la media de la columna 'mass_kg'
df['mass_kg'].fillna(df_imputed, 
                     inplace=True)
df.head(5)
#Adicional: df.to_csv('data/datos_imputados.csv')# para exportar el DatFrame en csv si se desea ver en una hoja de calculo u otro software
```

Luego se procede con volver a graficar los datos con un histograma, donde aumenta de 17 a 17.5 la frecuencia de los datos en los pesos de barcos o naves.

```python
plt.figure(figsize=(8, 6))# tamaño de la figura
plt.hist(df["mass_kg"], bins=20, edgecolor='black')
plt.xlabel("Pesos de las naves o barcos (kg)")
plt.ylabel("Frecuencia de datos")
plt.title("Grafico No. 2 - Histograma de los pesos de las naves 'mass_kg' imputados")
plt.grid(True)# añadir rejillas al grafico para una mejor interpretación de los datos
plt.show()
```
