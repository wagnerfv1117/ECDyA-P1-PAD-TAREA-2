Se procede con desarrollar las acciones pertinentes a extraer la información y desarrollar el punto conforme a lo estipulado en el material de Idris (2014); es importante resaltar que este procedimiento que se mostrará en la siguiente celda de código, es una API "interfaz de programación de aplicaciones" por sus siglas en inglés, donde se extrae la información de una pagina web Roldán (2019, p.72)

```python
import requests as rq# importar la librería request

data = rq.request('GET','https://api.spacexdata.com/v4/ships')#obtención del datos a través del método GET

data

to_json = data.json()# conversión de los datos extraidos a formato diccionario

to_json
```

<p style="text-align: justify;">En el siguiente código, se utiliza la función pd.DataFrame para crear un dataframe a partir de la lista de diccionarios devuelta por la API de SpaceX, luego se selecciona solo las columnas especíque pide la guia de aprendizaje y evaluación.</p>

```python
import pandas as pd
df = pd.DataFrame(to_json)[['type', 'mass_kg', 'year_built', 'home_port', 'name']]#extracción de las columnas necesarias
df
```

En el caso desglosado anteriormente, se observa la que hay varias columnas con valores faltantes o nulos, que pueden ser imputados, los cuales son representados con la palabra NaN, que son las siglas de Not a Number, que traduciendolo al español es no hay numero. Madhavan (2015)

```python
datos_faltantes = df['mass_kg'].isnull().sum()#revisión de los datos faltantes en la variable "mass_kg"
print("Los datos faltantes que hay en la columna 'mass_kg' es de:",datos_faltantes, 'registros')
```

Se desarrolla el DataFrame "df1", removiendo las filas que tienen algunos datos faltantes, utilizando el método dropna(), consultado en el libro de McKinney(2023), para este caso si como profesionales de datos, deseamos eliminar datos.

```python
df1 = df.dropna(subset=['mass_kg'])
```

Se desarrolla el DataFrame "df2", removiendo las filas que tienen algunos datos faltantes, utilizando el método dropna()

```python
df2 = df.dropna(how='all')
```

Se procede con completar los valores faltantes con el promedio de la columna con el metodo fillna Madhavan (2015)

```python
df.fillna(0, inplace=True)# el valor a insertar en los dataframe este 0
df.fillna(method='ffill', inplace=True)#se rellenan los valores faltantes utilizando el método de "forward fill" o "relleno hacia adelante".
```

Se procede con la imputación de los datos en columnas en blanco, donde se inserta el numero 0

```python
promedio = df['mass_kg'].mean()# metodo .mean() o el promedio de los datos
df['mass_kg'].fillna(promedio, inplace=True)
```

Se utiliza el metodo shape, para determinar la dimensión en columnas y filas del DataFrame, que es una función dentro de la librería Pandas que devuelve el número de filas y columnas de un array.

```python
df1 = df1.shape# se define el dataframe 1
df2 = df2.shape# se define el dataframe 2
print("La dimensión de los DataFrames, se puede observar así:")
print(f"df1: {df1[0]} filas, {df1[1]} columnas")# cero para filas y 1 para columnas
print(f"df2: {df2[0]} filas, {df2[1]} columnas")
```

Se procede con extraer la primera palabra de cada dato en la columna "home_port" y mostrar los resultados, utilizando el método str.split(), éste metodo se utiliza para dividir una columna de cadenas en un DataFrame en función de un delimitador, para después acceder al primer elemento de cada resultado. (Roldán, 2019)

```python
df['first_word'] = df['home_port'].str.split().str[0]#metodo para extraer la primera palabra de la columna
print(df['first_word'])
```

Se procede hacer  el filtrado de los datos que empienzan con las palabras "port", con palabras diferentes, y la palabra diferente como tal, para este caso, se utiliza el metodo startswith() que utiliza parámetros en: valor, comienzo y fin; es decir si comienza con, en qué posición está y en qué posición está la busqueda. Madhavan (2015)

```python
data_with_port = df[df['home_port'].str.startswith('Port')].shape[0]
data_without_port = df[df['home_port'].str.startswith('Port') == False].shape[0]
different_words = df[df['home_port'].str.startswith('Port') == False]['first_word'].unique()

"""Se utilizó el método str.startswith(), para verificar si una cadena comienza con otra cadena específica. 
    Para este caso, se utiliza en procesamiento de texto y extraer información de las columnas en este la palabra "port" """

print("Cantidad de datos que comienzan con 'Port'es de:", data_with_port)
print("Cantidad de datos que comienzan con una palabra diferente es de:", data_without_port)
print("Palabras diferentes que hay en el DataFrame es :", different_words)
```

Filtrar el DataFrame con los datos que en la columna name contengan en el string la palabra “GO” , aplicando el metodo string.contains, que ayuda a filtrar datos, consultandolo en el material de Idris (2024) 

```python
df_filtered = df[df['name'].str.contains('GO')]# aplicacion del metodo .str.contains para extraer la palabra GO
df_filtered
```

Convertir strings de la columna "name" a mayúsculas utilizando el metodo .str.upper() de la librería Pandas, que sirve para convertir todas las letras minúsculas de una cadena a mayúsculas.

```python
df['name'] = df['name'].str.upper()
print(df['name'])
```

Convertir strings de la columna "type" a minúsculas utilizando el metodo .str.lower(), que sirve para sirve para convertir todas las letras minúsculas de una cadena a mayúsculas. McKinney (2023)

```python
df["type"] = df["type"].str.lower()
print(df["type"])
```

<p style="text-align: justify;">Se procede con dividir todos los strings de la columna home_port por el espacio en blanco utilizando el metodo .str.split() que es un para dividir  las letras de la columna en cadenas de listas con su separador definido, facilitando el análisis y manipulación de texto Madhavan (2015)</p>

```python

df["port_parts"] = df["home_port"].str.split()

print(df[["home_port", "port_parts"]])
```
