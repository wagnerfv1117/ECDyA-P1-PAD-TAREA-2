```python
import pandas as pd
```

```python
concatened = pd.concat([concatened, p1])# p1 representa el DataFrame 1 el cual es el que se va a mostrar.
concatened
```

* Concatenar los DtaFrame p1 y p2 con sus respectivos datos

```python
concatened = pd.concat([p1, p2], keys=['p1', 'p2'])# asignacipn de llaves o claves para identificarlos similar a SQL
print("Concatenación con identificación de origen:")
print(concatened)
```

* Encontrar el número promedio de estudiantes que son obesos en escuelas de primaria (tipo ‘ELEMENTARY’).

```python
elementary_schools = df[df['GRADE LEVEL'] == 'ELEMENTARY']
elementary_schools#impresión del DataFrame
```

* Encontrar el número total de estudiantes de primaria que son obesos, que generalmente en Pandas para extraer la media, se usa el comando o metodo .mean() para calcular la media de esa columna, consultandolo en el material de Idris(2014)

```python
avg_obese_students = elementary_schools['NO. OBESE'].mean()#metodo mean que es la media
print(f"Promedio de estudiantes obesos en escuelas de primaria es de: {avg_obese_students}")
```

* Encontrar el número máximo de estudiantes obsesos que tiene una escuela, que consiste en hacer una sumatoria con el metodo .sum() quie conmsiste en hacer una suma general de los datos que tiene la columna 'NO. OBESE'

```python
total_obese_students = elementary_schools['NO. OBESE'].sum()#metodo sum() que adiciona todos los datos
print(f"Número total de estudiantes de primaria que son obesos es de : {total_obese_students}")
```

* Encontrar el número máximo de estudiantes obsesos que tiene una escuela, aplicando el metodo max() que consiste en busca el numero maximo que hay en la columna que se va analizar

```python
max_obese_students = elementary_schools['NO. OBESE'].max()#función max() para buscar el numero grande
print(f"Número mínimo de estudiantes obesos que tiene una escuela es de: {max_obese_students}")
```

* Encontrar el número mínimo de estudiantes obsesos que tiene una escuela, utilizando el metodo min() quie sirve para econtrar el valor minimo en la columna

```python
min_obese_students = elementary_schools['NO. OBESE'].min()#funcion min() para encontrar el numero minimo
print(f"Número mínimo de estudiantes obesos que tiene una escuela es de: {min_obese_students}")
```

* Obtener la desviación estándar del número de estudiantes obsesos en las escuelas, utilizando el metodo std() la cual cuantifica la dispersión o variabilidad de un conjunto de datos, para este caso es el DataFrame llamado "df" en la columna 'NO. OBESE'

```python
std_obese_students = elementary_schools['NO. OBESE'].std()# metodo std() para calcula la desviación estandar
print(f"Desviación estándar del número de estudiantes obesos en las escuelas es de: {std_obese_students}")
```

* Obtener la cantidad de escuelas de primarias en el condado (county) de DELAWARE.

```python
delaware_elementary_schools_count = elementary_schools[elementary_schools['COUNTY'] == 'DELAWARE'].shape[0]
""" aplicación del metodo shape[] para obtener las dimensiones de un arreglo o DataFrame """
print(f"La cantidad de escuelas de primaria en el condado de DELAWARE es de: {delaware_elementary_schools_count}")
```

Resultados y conclusiones: Los resultados al momento de aplicar lois diferentes metodos de la librería Pandas, son muy variados, concluyendo que utilizar tecnicas estadisticas, puede contribuir a tener un panorama amplio para comprender el comportamiento o tendencias en los datos.

* Encontrar la suma del número de estudiantes obesos en cada uno de los grados mediante el uso de la función groupby de pandas. Muestre sus resultados y explíquelo.

```python
sum_obese_by_grade = df.groupby('GRADE LEVEL')['NO. OBESE'].sum().reset_index()# metodo sum() para sumar y reset_index() para resetear indices

sum_obese_by_grade
```

* Medidas de tendencia central de estudiantes obesos en cada nivel estuantil, aplicando los métodos 'sum', 'mean', 'std' integrados en la librería Pandas Idris (2014)

```python
agg_obese_by_grade = df.groupby('GRADE LEVEL')['NO. OBESE'].agg(['sum', 'mean', 'std']).reset_index()# se aplica los metodos para calcular medidas de tendencia central

print("Agregación por suma, media y desviación estándar del número de estudiantes obesos en cada nivel:")
print(agg_obese_by_grade)
```
